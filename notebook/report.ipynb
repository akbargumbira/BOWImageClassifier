```python
plt.plot(x, np.log(silhouette_index))
plt.title('Silhoutte Index')
plt.show()

plt.plot(x, np.log(calinski_harabaz_index))
plt.title('Calinski Harabaz Index')
plt.show()
```

```python
print(silhouette_index)
print np.diff(x)
gradient = np.diff(silhouette_index)/np.diff(x)
print gradient
print("The most stable one %s, %s, %s") % (np.argmax(gradient), x[np.argmax(gradient)], x[np.argmax(gradient)+1])
print("The steepest one %s, %s, %s") % (np.argmin(gradient), x[np.argmin(gradient)], x[np.argmin(gradient)+1])
```
